```python
def TeamStat(team):
    home_goal_positiv = 0 #Голов забито дома
    home_goal_negativ = 0 #Голов пропущено дома
    home_win = 0 #Выиграно дома
    home_draw = 0 #Ничья дома
    home_lost = 0 #Проиграно дома
    home_score = 0 #Количество набранных очков дома
    home_matches = 0 #Количество сыгранных матчей дома
    home_shot_positiv = 0 #Удары команды дома
    home_shot_negativ = 0 #Удары соперника дома
    
    away_goal_positiv = 0 #Голов забито в гостях
    away_goal_negativ = 0 #Голов пропущено в гостях
    away_win = 0 #Выиграно в гостях
    away_draw = 0 #Ничья в гостях 
    away_lost = 0 #Проиграно в гостях
    away_score = 0 #Количество набранных очков в гостях
    away_matches = 0 #Количество сыгранных матчей в гостях
    away_shot_positiv = 0 #Удары команды в гостях
    away_shot_negativ = 0 #Удары соперника в гостях
    
    all_goal_positiv = 0 #Голов забито
    all_goal_negativ = 0 #Голов пропущено
    all_win = 0 #Выиграно
    all_draw = 0 #Ничья
    all_lost = 0 #Проиграно
    all_score = 0 #Количество набранных очков
    all_matches = 0 #Количество сыгранных матчей дома
    all_shot_positiv = 0 #Удары команды
    all_shot_negativ = 0 #Удары соперника
    
    for i in range(len(df_RPL)):
        if df_RPL['home_team'][i] == team :
            home_matches += 1
                
            home_goal_positiv += df_RPL['home_goals'][i]
            home_goal_negativ += df_RPL['away_goals'][i]

            if (df_RPL['home_goals'][i] > df_RPL['away_goals'][i]):
                home_score += 3
                home_win += 1
            elif (df_RPL['home_goals'][i] < df_RPL['away_goals'][i]):
                home_lost +=1
            else:
                home_score += 1
                home_draw += 1
            
            home_shot_positiv += df_RPL['home_shots'][i]
            home_shot_negativ += df_RPL['away_shots'][i]
        
        if df_RPL['away_team'][i] == team :
            away_matches += 1
                
            away_goal_positiv += df_RPL['away_goals'][i]
            away_goal_negativ += df_RPL['home_goals'][i]

            if (df_RPL['away_goals'][i] > df_RPL['home_goals'][i]):
                away_score += 3
                away_win += 1
            elif (df_RPL['away_goals'][i] < df_RPL['home_goals'][i]):
                away_lost +=1
            else:
                away_score += 1
                away_draw += 1
            
            away_shot_positiv += df_RPL['away_shots'][i]
            away_shot_negativ += df_RPL['home_shots'][i]    
            
        all_goal_positiv = home_goal_positiv + away_goal_positiv   
        all_goal_negativ = home_goal_negativ + away_goal_negativ
        all_win = home_win + away_win
        all_draw = home_draw + away_draw
        all_lost = home_lost + away_lost
        all_score = home_score + away_score
        all_matches = home_matches + away_matches
        all_shot_positiv = home_shot_positiv + away_shot_positiv
        all_shot_negativ = home_shot_negativ + away_shot_negativ
    
    return [home_goal_positiv, home_goal_negativ, home_win,
            home_draw, home_lost, home_score, home_matches,
            home_shot_positiv, home_shot_negativ,
            away_goal_positiv, away_goal_negativ, away_win,
            away_draw, away_lost, away_score, away_matches,
            away_shot_positiv, away_shot_negativ,
            all_goal_positiv, all_goal_negativ, all_win,
            all_draw, all_lost, all_score, all_matches,
            all_shot_positiv, all_shot_negativ]
```

```python
home_team_vectors = collections.defaultdict(list)
for team in df_RPL.home_team.unique():
        home_team_vector = homeTeamStat(team)
        home_team_vectors[team] = home_team_vector
```

```python
away_team_vectors = collections.defaultdict(list)
for team in df_RPL.home_team.unique():
        away_team_vector = awayTeamStat(team)
        away_team_vectors[team] = away_team_vector
```

```python
df_RPL_train = df_RPL[:-10]
df_RPL_test = df_RPL[-10:]
```

```python
df_RPL_test
```

```python
xTrain = np.zeros(( len(df_RPL_train.index), len(homeTeamStat('Уфа'))))
yTrain = np.zeros( len(df_RPL_train.index) )
yyTrain = np.zeros(( len(df_RPL_train.index) ))
yTrain_shot = np.zeros( len(df_RPL_train.index) )
yyTrain_shot = np.zeros(( len(df_RPL_train.index) ))
counter = 0
for index, row in df_RPL_train.iterrows():
    team = row['home_team']
    t_vector = home_team_vectors[team]
    rivals = row['away_team']
    r_vector = away_team_vectors[rivals]
           
    diff = [a - b for a, b in zip(t_vector, r_vector)]
            
    if len(diff) != 0:
        xTrain[counter] = diff
    if team == row['winner']:
        yTrain[counter] = 1
    else: 
        yTrain[counter] = 0
    
    if rivals == row['winner']:
        yyTrain[counter] = 1
    else: 
        yyTrain[counter] = 0
    yTrain_shot[counter] = row['home_shots']
    yyTrain_shot[counter] = row['away_shots']
    counter += 1     
```

```python
xTest = np.zeros(( len(df_RPL_test.index), len(homeTeamStat('Уфа'))))
counter = 0
for index, row in df_RPL_test.iterrows():
    team = row['home_team']
    t_vector = home_team_vectors[team]
    rivals = row['away_team']
    r_vector = away_team_vectors[rivals]
           
    diff = [a - b for a, b in zip(t_vector, r_vector)]
            
    if len(diff) != 0:
        xTest[counter] = diff
   
    counter += 1     
```

```python
model1 = LinearRegression()
model2 = LinearRegression() 
xg_model1 = XGBRegressor()
xg_model2 = XGBRegressor()
model1.fit(xTrain, yTrain)
model2.fit(xTrain, yyTrain)
xg_model1.fit(xTrain, yTrain)
xg_model2.fit(xTrain, yyTrain)

model1_shot = LinearRegression()
model2_shot = LinearRegression() 
xg_model1_shot = XGBRegressor()
xg_model2_shot = XGBRegressor()
model1_shot.fit(xTrain, yTrain_shot)
model2_shot.fit(xTrain, yyTrain_shot)
xg_model1_shot.fit(xTrain, yTrain_shot)
xg_model2_shot.fit(xTrain, yyTrain_shot)
```

```python
print(model1.predict(xTest))
print(model2.predict(xTest))
print(xg_model1.predict(xTest))
print(xg_model2.predict(xTest))
```

```python
df_RPL_test['вероятность победы home_team(усреднение моделей)'] = (xg_model1.predict(xTest)+model1.predict(xTest))/2
df_RPL_test['вероятность победы away_team(усреднение моделей)'] = (xg_model2.predict(xTest)+model2.predict(xTest))/2
df_RPL_test['прогнозируемое число ударов home_team'] = (xg_model1_shot.predict(xTest)+model1_shot.predict(xTest))/2
df_RPL_test['прогнозируемое число ударов away_team'] = (xg_model2_shot.predict(xTest)+model2_shot.predict(xTest))/2
```

```python
rpl_final = df_RPL_test.drop(['index','date', 'championship', 'home_goals',
                  'away_goals'] , axis=1)
```
